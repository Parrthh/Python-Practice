```python
#Q20 Make a deep copy of list and check whether it is copying deeply or not
m = [1,44,89]
z = m
#Printing the new list
print(z)
#By making changes in the original list we can see that they have been reflected in the new list as well
m.append(100)
print(m)
print(z)
```

```python
#Q21 Make a copy of list in that way  making changes in one list will not affect to the another list
a = ['six','seven','eight']
b = a.copy()
print(b)
#Now even if I make changes in 'a' variable the changes wont affect the 'b' variable
a.append('five')
print(a)
print(b)
```

```python
#Q22 Take any example to  explain stack and queue
#Stack Example:
a = ['India','UK','USA']
a.append('Canada')
print(a)
a.pop()
print(a)

#Queue Example
b = ['India','UK','USA']
b.append('Canada')
print(b)
b.pop(0)
print(b)
```

```python
#Q23 Zip four list contain with natural number your name,age, roll no
a = ['Parth']
b = [21]
c = [37]
#While using zip we need to take a new variable to store the zip contents and then convert it into a list'
d = zip(a,b,c)
d = list(d)
print(d)
```

```python
#Q24 Unzip the above list
a, b, c = zip(*d)
print(a)
print(b)
print(c)
```
