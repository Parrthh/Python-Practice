## OPERATORS

### Arithmetic Operator

```python
#Addition
x = 10.6
y = 50
print(x+y)
```

```python
#Subtraction
x = 50
y = 100
print(x-y)
```

```python
#Multiplication
x = 19
y = 77
print(x*y)
```

```python
#Division
x = 52
y = 4
print(x/y)
```

```python
#Floor Division (It returns the quotient (the result of division) in which the digits after the decimal are not printed)
x = 7.99
y = 2
#Normal Division
print(x/y)
#Floor Division - For floor division you need to use '//'
print(x//y)
```

```python
#Modulus - gives us the remainder of the division
#For Modulus you need to use '%'
x = 77 
y = 6 
print(x%y)
```

```python
#Exponent - power of a number
num = 5
#While using exponent you need to write in this way 'num**2'
#Here I am writing 2 as I need the square of the number
print(num**2)
```
