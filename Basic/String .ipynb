### String

```python
s = 'String in python'
print(s)
```

```python
len(s)
```

```python
#Multi line strings can be written under triple quotes
a = '''Hello
My name is XYZ
I am __ years old'''
print(a)
```

```python
#Positive indexing
b = 'String in Python'
b[10]
```

```python
#Negative Indexing
c = 'This is a string'
c[-8]
```

```python
#String is unmutable hence it is not possible to modify it
c = 'This is a string'
c[2] = a
print(c)
```

```python
c = 'This is a string'
c[1:6]
```

```python
c = 'This is a string'
c[1:9:4]
```

```python
c = 'This is a string'
c[:-1]
```

```python
c = 'This is a string'
c[-7:-1]
```

### String Function

#### 1. Lower()

```python
a = 'String in Python'
a.lower()
```

#### 2. Upper()

```python
a = 'String in Python'
a.upper()
```

#### 3. Title()

```python
a = 'string in python'
a.title()
```

#### 4. Capitalize()

```python
a = 'sTring in pYthon'
a.capitalize()
```

#### 5. Casefold

```python
a = 'String in Python'
a.casefold()
```

#### 6. Split

```python
a = 'String in Python'
a.split()
```
